
### Featuring Engineering

#### We will split data into train and test data to prevent data leakage

```python
X_train,X_test,y_train,y_test=train_test_split(data,data['SalePrice'],test_size=0.3,random_state=42)
print(X_train.shape)
print(X_test.shape)
                                              
```

```python
# we will store all categorical features with missing features
missing_cat=[feat for feat in categorical if data[feat].isna().sum()>0 and data[feat].dtype=='O']
missing_cat
```

```python
# Checking missing values for categorical variables
for i in missing_cat:
    print("{}% of missing features in {}".format((data[i].isna().sum()*100/len(data[i])),i))
```

```python
# we will replace missing values with a new label by creating a function
def fillmissing(db,miss):
    data=db.copy()
    data[miss]=data[miss].fillna('missing')
    return data

X_train=fillmissing(X_train,missing_cat)
    
```

```python
X_train[missing_cat].isna().sum()  # no null values in categorical columns in train data
```

```python
# checking null values in numerical variables
numerical_nan=[a for a in numerical_features if X_train[a].isna().sum()>1]
```

```python
numerical_nan # there are only 3 of them with missing values
```

```python
for i in numerical_nan:
    print('the percentage of missing values in {} is {}'.format(i,(X_train[i].isna().sum()*100/len(X_train[i]))))
          
```

```python
# since there are outliers we will impute missing values with median
for i in numerical_nan:
    median=X_train[i].median()
    X_train[i+'nan']=np.where(X_train[i].isna(),1,0) # preserving the columns having null values to keep information
    X_train[i].fillna(median,inplace=True)
X_train[numerical_nan].isna().sum()   
```

#### We are going to convert the temporal variables to total years lapsed

```python
features
```

```python
for i in features:
    if i!='YrSold':
        X_train[i]=X_train['YrSold']-X_train[i]   
```

```python
X_train[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head(10)
```

#### removing skewness from continuous numerical variables through log transformation}

```python
# We will now transform the continuous numerical features
# as log of 0 is undefined so we need to take only nonzero columns for log transformation
cont_without_0=[i for i in cont if 0 not in X_train[i].unique()]
```

```python
cont_without_0 # These columns do not have 0 in them.
```

```python
for i in cont_without_0:
    X_train[i]=np.log(X_train[i])
```

```python
X_train[cont_without_0].head()
```

#### We will handle rare categorical features. 

```python
# we will remove those features which have very less percentage count in that column. It does not provide significant variance to data
for i in categorical:
    temp_df=X_train.groupby(i)['SalePrice'].count()/len(X_train)
    temp_df=temp_df[temp_df>0.01].index  # getting index of the rows with less than 0.01% of count
    X_train[i]=np.where(X_train[i].isin(temp_df),X_train[i],'Rare')
```
